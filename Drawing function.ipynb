```python
import pylab
import numpy

x = numpy.linspace(0,5,100) # 100 linearly spaced numbers
y = x ** 2 # computing the values of x^2
z = 3 ** x
a = x ** 3
# compose plot
pylab.plot(x,y) 
pylab.plot(x,z)
pylab.plot(x,a)
```

```python
# Import our modules that we are using
import matplotlib.pyplot as plt
import numpy as np

# Create the vectors X and Y
x = np.array(range(100))
y = x ** 2

# Create the plot
plt.plot(x,y)

# Show the plot
plt.show()
```
