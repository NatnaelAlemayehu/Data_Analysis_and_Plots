```python
import pylab
import numpy as np
from scipy import optimize
from scipy.optimize import minimize

x = np.linspace(-10,10,100) # 100 linearly spaced numbers
y = np.sin(x)*np.exp(-0.1*(x-0.6)**2) # computing the values of x^2

# compose plot
pylab.plot(x,y)

```

```python
def scalar1(x):
    return np.sin(x)*np.exp(-0.1*(x-0.6)**2)
result = optimize.minimize_scalar(scalar1)
print(result)
print("Minimum occurs at: ",result['x'])
max_x = minimize(lambda x: -scalar1(x), 0)
print("Maximum occurs at: ", max_x.x)
```
